```r
linearMod <- lm(perm ~ peri, data=rock)  # am construit un model de regresie liniara pe perimetrul si permeabilitatea
print(linearMod)
print(summary(linearMod))
with(rock,plot(peri, perm))
	abline(linearMod)
AIC(linearMod)
BIC(linearMod)
```

```r
generated_data <- rnorm(rock$area,mean(rock$area),var(rock$area))
rock$generated_area <- generated_data
```

```r
par(mfrow=c(1, 2))
qqnorm(rock$area, pch = 1, frame = FALSE)
qqline(rock$area, col = "steelblue", lwd = 2)

qqnorm(rock$generated_area, pch = 1, frame = FALSE)
qqline(rock$generated_area, col = "steelblue", lwd = 2)
# se observa in q-q plot ca datele generate respecta aproximativ aceeasi distributie
```

```r
linearMod <- lm(perm ~ area+generated_area, data=rock)  # am construit un model de regresie liniara care se foloseste de shape
print(linearMod)
print(summary(linearMod))
```
